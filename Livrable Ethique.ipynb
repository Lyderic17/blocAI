<h3 style="text-decoration: underline;  text-weight: bold">But de l'étude</h3>

<p>Le but est de trouver la ou les raisons pour laquelle les employés seraient enclins à démissionner.</p>
<p>Les résultats seront donnés aux ressources humaines afin de savoir quels changements devraient être réalisés afin de garder le plus d’employés possibles le plus longtemps possible.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Contexte</h3>

<p>L'entreprise de produits pharmaceutiques HumanForYou basée en Inde emploie environ 4000 personnes. Cependant, chaque année elle subit un turn-over d'environ 15% de ses employés nécessitant de retrouver des profils similaires sur le marché de l'emploi.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Problème</h3>

<p>Le problème est de définir quelles données sont intéressantes à étudier car tout ne peut pas être pris en compte.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Solution</h3>

<p>La solution est de réaliser un modèle d’intelligence artificielle qui mettrait en relation les différentes données afin de trouver une solution et ou une piste de solution afin de rendre nul ou au moins de réduire le taux de démission</p>

<h3 style="text-decoration: underline;  text-weight: bold">Alternatives possibles</h3>

<p>Prendre les données brutes et en faire des déductions. La limite à cette solution est qu’elle pourrait donner lieu à des erreurs d’interprétation et des conclusions faites trop vite.</p>
<p>(exemple : Plusieurs personnes vieilles partent de l’entreprise. La déduction première serait de dire, ils sont vieux, on les change. Or, peut être que ces données sont à corréler avec d’autres données, tels que, le manager n’est pas bon. Donc utiliser la donnée brute serait plus destructeur pour l’entreprise car cela signifierai perdre une personne de valeur pour l’entreprise.)</p>
<p>Il faut donc réaliser un modèle d’intelligence artificielle.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Parties prenantes, personnes impactées</h3>

<p>Avec de mauvaises conclusion et déductions, les personnes impactées seraient les employés dont les données ont été mal étudiées, mais également les ressources humaines et à fortiori l’entreprise, car elle perdraient des personnes qui ne comptaient pas partir ou dont le travail était suffisant.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Non éthique lié à l’IA</h3>

<p>Les biais de l’IA, mauvais calibrage et intégration de données mauvaises. ( cf après.)</p>
<p>Cela impacterait également les futures recrues, car si l’on se base faussement sur certaines données , certains employés pourraient avoir des caractéristiques qui correspondent aux fausses conclusions, et donc par conséquent, seraient plus enclins à quitter l’entreprise. Donc ce n’est pas bon.</p>
<p>Avec de meilleurs conclusions et déductions ( donc un modèle d’intelligence artificielle) les personnes impactées seraient les ressources humaines, car cela influencerai dans le bon sens toute conclusion à faire, et également les futures personnes à recruter, qui quitteraient moins l’entreprise.</p>

<h3 style="text-decoration: underline;  text-weight: bold">Filtre éthique</h3>

<p>Le premier filtre éthique serait de ne pas renvoyer une personne sans raison en se basant sur des données sans demander l’avis de la personne ou son ressenti.</p>

<h4 style="text-weight: bold">L’éducation :</h4>

<p>Si l’on se base sur les études que quelqu’un a fait, cela serait discriminant, et ne serait pas éthique car ne laisserai pas beaucoup de chance aux personnes qui ont fait peu d’étude, donc le système deviendrait élitiste.</p>

<h4 style="text-weight: bold">L’âge :</h4>

<p>L’âge est une donnée sur laquelle nous ne devrions pas nous baser INDIVIDUELLEMENT, mais elle sera prise dans le contexte de ce projet avec d’autres données, car l’âge peut nous permettre de faire des corrélations avec d’autres données telles que le nombre d’années dans l’entreprise, l’expérience de la personne, la dernière promotion reçue ( comparé à leur âge, la dernière promotion reçue.)<p>
<p>Donc dans notre cas l’âge est intéressant à utiliser, et éthique tant que l’on ne prend pas QUE l’âge.</p>

<h4 style="text-weight: bold">Statut marital :</h4>

<p>Si l’on souhaitait utiliser le statut marital, il faudrait l’utiliser seulement pour savoir le nombre de congés parental. Le but de cette donnée serait seulement de prédire si la personne pourrait partir en congé parental. Mais en dehors de cela, il ne serait pas intéressant car on peut être marié mais sans enfant, sans envie de faire des enfants etc.. Ce paramètre est trop aléatoire. L’inclure dans le choix de renvoie d’une personne serait également fortement déplacé.</p>

<h4 style="text-weight: bold">Stock Option Level :</h4>

<p>Le fait d’avoir des parts dans l’entreprise ne doit pas influer sur la décision finale car cela voudrait que plus quelqu’un détient de part, moins elle serait en danger d’être renvoyée.</p>

<h4 style="text-weight: bold">Genre :</h4>

<p>Ce paramètre n’est pas éthique car on ne peut baser une décision sur le genre d’une personne.</p>


<h3 style="text-decoration: underline;  text-weight: bold">Données inutilisables</h3>

<p>Certaines données sont inutiles à utiliser car ne contiennent que la même valeur ( cf. Over18, ou tout le monde a forcément plus de 18 ans, donc la donnée est inutilisable et inutile à des fins d’interprétation.)</p>

<h4 style="text-weight: bold">Employee count :</h4>

<p>Est inutile car nous n’aurons pas son âge car il arrive après 2015, et la donnée en elle-même n’est pas utile.</p>

<h4 style="text-weight: bold">Standard hours :</h4>

<p>Les données sont également toujours les mêmes, donc ce paramètre ne permet pas de différencier et d’apporter quelque chose de valeur dans notre modèle.</p>
