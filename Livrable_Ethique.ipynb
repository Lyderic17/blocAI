<h2>I. Introduction</h2>
<p>Le 17 juillet 2020, le groupe d'experts de haut niveau sur l'intelligence artificielle (AI HLEG) a présenté sa liste d'évaluation finale pour une intelligence artificielle digne de confiance.</p>

<h2>II. Développement de notre plan éthique</h2>
<h3>1. Situation initiale</h3>
<h4 style="text-decoration: underline;  text-weight: bold">But de l'étude</h4>

<p>Le but est de trouver la ou les raisons pour laquelle les employés seraient enclins à démissionner.</p>
<p>Les résultats seront donnés aux ressources humaines afin de savoir quels changements devraient être réalisés afin de garder le plus d’employés possibles le plus longtemps possible.</p>

<h4 style="text-decoration: underline;  text-weight: bold">Contexte de l'étude</h4>

<p>L'entreprise de produits pharmaceutiques HumanForYou basée en Inde emploie environ 4000 personnes. Cependant, chaque année elle subit un turn-over d'environ 15% de ses employés nécessitant de retrouver des profils similaires sur le marché de l'emploi.</p>

<h4 style="text-decoration: underline;  text-weight: bold">Problème</h4>

<p>Le problème est de définir quelles données sont intéressantes à étudier car tout ne peut pas être pris en compte, compte tenu de la problématique de l'éthique et de la protection des données.</p>

<h3>2. Solution</h3>
<h4 style="text-decoration: underline;  text-weight: bold">Alternatives possibles</h4>

<p>À partir des consignes posées par le groupe d'experts de haut niveau sur l'intelligence artificielle, nous devons définir quelles données sont utilisables de manière éthiques sans re-créer les biais présents dans les consciences humaines mais aussi exploitables dans notre modèle utilisant une intelligence artificielle permettant de prospecter de futurs départs.</p>

<p>(exemple : Plusieurs personnes vieilles partent de l’entreprise. La déduction première serait de dire, ils sont vieux, on les change. Or, peut être que ces données sont à corréler avec d’autres données, tels que, le manager n’est pas bon. Donc utiliser la donnée brute serait plus destructeur pour l’entreprise car cela signifierai perdre une personne de valeur pour l’entreprise.)</p>
<p>Il faut donc réaliser un modèle d’intelligence artificielle.</p>

<h3>3. Parties prenantes</h3>
<h4 style="text-decoration: underline;  text-weight: bold">Personnes impactées</h4>
<p>Avec de mauvaises conclusion et déductions, les personnes impactées seraient les employés mal à l'aise dans l'entreprise, mais également les ressources humaines et à fortiori l’entreprise, car elle perdraient des personnes qui ne comptaient pas partir ou dont le travail était suffisant.</p>
<p>On a donc trois principales personnes impactées :</p>
<li style="list-style-type: none; margin-bottom: 5px; margin-top: 5px">- Les employés</li>
<li style="list-style-type: none; margin-bottom: 5px">- Les ressources humaines</li>
<li style="list-style-type: none">- L'entreprise</li>

<h4 style="text-decoration: underline;  text-weight: bold">Moyen à mettre en place pour prévenir/réduire les inconvénients</h4>

<p>Limiter les biais de l’IA, mauvais calibrage et intégration de mauvaises données. Ce qui supprimerait des profils intéressant pour l'entreprise. Cela impacterait également les futures recrues, car si l’on se base faussement sur certaines données, certains employés pourraient avoir des caractéristiques qui correspondent aux fausses conclusions, et donc par conséquent, seraient plus enclins à quitter l’entreprise.</p>
<p>Restreindre l'utilisation des données de performance des employés pour ne par rajouter une nouvelle pression qui les pousseraient à démissionner.</p>
<p>Remettre en question les conclusions de l'outil et ne pas boire ses conclusions trop hativement.</p>
<p>Créer une démarche de discussion entre les employés et le service des ressources humaines, afin d'exprimer des besoins ou bien des problèmes qui pourraient créer des mal êtres au sein de l'entreprise</p>

<h3>5. Filtre éthique</h3>

<p>Ce filtre éthique permettra de ne pas renvoyer une personne sans raison en se basant sur des données non basées sur une chartre éthique. Les différences sont aussi des forces et donc cette recherche de la raison des départs ne peut pas se baser sur des critères discriminants.</p>

<h4 style="text-weight: bold">L’éducation :</h4>

<p>Si l’on se base sur les études que quelqu’un a fait, cela serait discriminant, et ne serait pas éthique car ne laisserai pas beaucoup de chance aux personnes qui ont fait peu d’étude, donc le système deviendrait élitiste. Même chause pour la donnée EducationField.</p>
<p>On peut lier cette donnée à la diversité, non-discrimination et équité.</p>

<h4 style="text-weight: bold">L’âge :</h4>

<p>L’âge est une donnée sur laquelle nous ne devrions pas nous baser INDIVIDUELLEMENT, mais elle sera prise dans le contexte de ce projet avec d’autres données, car l’âge peut nous permettre de faire des corrélations avec d’autres données telles que le nombre d’années dans l’entreprise, l’expérience de la personne, la dernière promotion reçue (comparé à leur âge, la dernière promotion reçue.)<p>
<p>Donc dans notre cas l’âge est intéressant à utiliser, et éthique tant que l’on ne prend pas QUE l’âge en compte.</p>
<p>On peut lier cette donnée à la protection de la vie privée et à la diversité, non-discrimination et équité.</p>

<h4 style="text-weight: bold">Statut marital :</h4>

<p>Si l’on souhaitait utiliser le statut marital, il faudrait l’utiliser seulement pour savoir le nombre de congés parental. Le but de cette donnée serait seulement de prédire si la personne pourrait partir en congé parental. Mais en dehors de cela, il ne serait pas intéressant car on peut être marié mais sans enfant, sans envie de faire des enfants etc.. Ce paramètre est trop aléatoire. L’inclure dans le choix de renvoie d’une personne serait également fortement déplacé.</p>
<p>On peut lier cette donnée à la protection de la vie privée.</p>

<h4 style="text-weight: bold">Stock Option Level :</h4>

<p>Le fait d’avoir des parts dans l’entreprise ne doit pas influer sur la décision finale car cela voudrait que plus quelqu’un détient de part, moins elle serait en danger d’être renvoyée.</p>
<p>On peut lier cette donnée à l'agencement humain et la surveillance.</p>

<h4 style="text-weight: bold">Genre :</h4>

<p>Ce paramètre n’est pas éthique car on ne peut baser une décision sur le genre d’une personne.</p>
<p>On peut lier cette donnée au bien-être environnemental et sociétal et responsabilité.</p>    

<h4 style="text-weight: bold">JobInvolvment :</h4>
    
<p>Comme son nom l'indique, cette donnée mesure l'impliquation de l'employé dans son travail. Utiliser cette donnée serait créer une pression suplémentaire sur les employés ce qui pourrait créer un encore plus grand nombre de turn over.</p>
<p>On peut lier cette donnée au bien-être environnemental et sociétal et responsabilité, à la transparence sur l'usage des données et à l'agencement humain et la surveillance.</p> 
    
<h3>6. Données inutilisables :</h3>

Certaines données sont inutiles à utiliser car ne contiennent que la même valeur :

<h4 style="text-weight: bold">Over18 :</h4>

<p>Dans notre jeu de donnée, tout le monde a plus de 18 ans, donc la donnée est inutilisable et inutile à des fins d’interprétation. Valeur booléenne pas vraiment utilisable.</p>

<h4 style="text-weight: bold">EmployeeCount :</h4>
    
<p>Inutile car pas de valeur réel dans l'interprétation et pas vraiment d'impact sur une cause de départ. La donnée YearatCompany peut la remplacer.</p>

<h4 style="text-weight: bold">StandardHours :</h4>
    
<p>Tout les employés travail le même nombre d'heure, donc ce paramètre ne permet pas de différencier et d’apporter quelque chose de valeur dans notre modèle.</p>
